```python
name = 'Свердловская область'
year = 2014
criterions = {}
```

```python
old_houses = pd.read_excel('Критерии.xlsx', sheet_name=2, skiprows=4).rename(columns={'Unnamed: 0': 'Название'})
old_houses
```

```python
old_houses[old_houses['Название'] == name]
```

```python
criterions['Доля площади многоквартирных домов, признанных аварийными, в общей площади многоквартирных домов'] = old_houses[old_houses['Название'] == name][year].iloc[0]
```

```python
children_pre_school = pd.read_excel('Критерии.xlsx', sheet_name=3).rename(columns={'Unnamed: 0': 'Название', 'Unnamed: 7': 'Доля'})
children_pre_school
```

```python
children_pre_school[children_pre_school['Название'] == 'муниципальное образование "город Екатеринбург"']
```

```python
criterions['Доля детей в возрасте 1 - 6 лет, состоящих на учете для определения в дошкольные образовательные учреждения, в общей численности детей в возрасте 1 - 6 лет (%)'] = children_pre_school[
                            children_pre_school['Название'] == 'муниципальное образование "город Екатеринбург"']['Доля'].iloc[0]
```

```python
accidents_people = pd.read_excel('Критерии.xlsx', sheet_name=4).rename(columns={'Unnamed: 15': 2014, 'Unnamed: 1': 'Всего'})
accidents_people.head(10)
```

```python
all_accidents = accidents_people[accidents_people.iloc[:, 0] == name]['Всего'].iloc[0]
all_people_of_name = accidents_people[year].iloc[2:].sum()

criterions['Количество дорожно-транспортных происшествий по отношению к численности населения в город (безразмерный коэффициент)'] = all_accidents / all_people_of_name
```

```python
people_with_water = pd.read_excel('Критерии.xlsx', sheet_name=5).rename(columns={'Unnamed: 1': 'Доля'})
people_with_water['Доля'] /= 100
people_with_water
```

```python
criterions['Доля городского населения, обеспеченного качественной питьевой водой из систем централизованного водоснабжения, в общей численности городского населения (процентов)'] = people_with_water[
                                                                            people_with_water.iloc[:, 0] == year]['Доля'].iloc[0]
```

```python
criterions
```

```python
name_criterions.append({'name': name,
                        'year': [{year: criterions}]})
```
