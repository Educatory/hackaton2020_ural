```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name_criterions = []
# [{'name': 'Свердловская область',
#   'year': [{'2014': {'Доля аварийного жилья': 0.6,
#      'Доля детей для определения в ДОУ': 0.4360968153,
#      'Доля населения, обеспеченного безопасной питьевой водой': 0.893831}}]}]
```

```python
name = 'Свердловская область'
year = 2014
criterions = {}
```

```python
old_houses = pd.read_excel('Критерии.xlsx', sheet_name=2, skiprows=4).rename(columns={'Unnamed: 0': 'Название'})
old_houses
```

```python
old_houses[old_houses['Название'] == name]
```

```python
criterions['Доля аварийного жилья'] = old_houses[old_houses['Название'] == name][year].iloc[0]
```

```python
children_pre_school = pd.read_excel('Критерии.xlsx', sheet_name=3).rename(columns={'Unnamed: 0': 'Название', 'Unnamed: 7': 'Доля'})
children_pre_school
```

```python
children_pre_school[children_pre_school['Название'] == 'муниципальное образование "город Екатеринбург"']
```

```python
criterions['Доля детей для определения в ДОУ'] = children_pre_school[
                            children_pre_school['Название'] == 'муниципальное образование "город Екатеринбург"']['Доля'].iloc[0]
```

```python
accidents_people = pd.read_excel('Критерии.xlsx', sheet_name=4).rename(columns={'Unnamed: 15': 2014, 'Unnamed: 1': 'Всего'})
accidents_people.head(10)
```

```python
all_accidents = accidents_people[accidents_people.iloc[:, 0] == name]['Всего'].iloc[0]
all_people_of_name = accidents_people[2014].iloc[2:].sum()

criterions['Доля ДТП к численности населения'] = all_accidents / all_people_of_name
```

```python
people_with_water = pd.read_excel('Критерии.xlsx', sheet_name=5).rename(columns={'Unnamed: 1': 'Доля'})
people_with_water['Доля'] /= 100
people_with_water
```

```python
criterions['Доля населения, обеспеченного безопасной питьевой водой'] = people_with_water[
                                                                            people_with_water.iloc[:, 0] == year]['Доля'].iloc[0]
```

```python
criterions
```

```python
name_criterions.append({'name': name,
                        'year': [{year: criterions}]})
```

```python
import json
```

```python
with open('criterions.json', 'w') as file:
    json.dump(name_criterions, file)
```
